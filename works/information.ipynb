# Task 2
Об'єднати багато малих рядків в один довгий рядок

```python
words = [
    'Листя', 'зелене', ',', 'птахи', 'співають', ',', 'а', 'все', 'навіть', 'квітне', 'навкруги', '.'
]

signs = [',', '.', '!', '?']
sentence = ''

for i in range(len(words)):
    sentence += words[i]
    # First, check if this is not the last element, and only then check the next one
    if i < len(words) - 1 and words[i+1] not in signs:
        sentence += ' '

print(sentence)
```

# Task 4
Додати 'd' днів до заданої дати (data) і вивести результат

```python
import datetime 

my_date  = datetime.datetime(2017, 6, 23, 10, 10, 10)
delta_days = 430

sentence = 'my {0:%B %d, %Y} be {0:%H:%M:%S}'.format(my_date)
print(sentence)

## add time period to my_date
new_date = my_date + datetime.timedelta(days=delta_days)

sentence = 'my {0:%B %d, %Y} be {0:%H:%M:%S}'.format(new_date)
print(sentence)
print(new_date.strftime('%m/%d/%Y'))
```
