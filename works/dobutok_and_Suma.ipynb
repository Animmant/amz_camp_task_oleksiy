## Problem Statement

**The Impossible Puzzle (Muddy Children Problem variant)**

Two players D and C are given two numbers `a` and `b` where `2 ≤ a,b ≤ 99`.
- D knows the **product** of the numbers
- C knows the **sum** of the numbers

The dialogue:
1. **D**: "I don't know the numbers"
2. **C**: "I knew you didn't know"
3. **D**: "Now I know the numbers"
4. **C**: "Now I know too"

**Goal**: Find all pairs (a, b) that satisfy this logical dialogue.

```python
# Initialize arrays and multiplication table

# Array of valid numbers: 2 to 99 inclusive
numbers = list(range(2, 100))
print(f"Valid numbers: {numbers[:5]}...{numbers[-5:]}")  # Show first 5 and last 5
print(f"Total numbers: {len(numbers)}")

# Generate multiplication table for numbers 2-99
multiplication_table = {}

for a in numbers:
    for b in numbers:
        if a <= b:  # Store only unique pairs (a <= b) to avoid duplicates
            product = a * b
            if product not in multiplication_table:
                multiplication_table[product] = []
            multiplication_table[product].append((a, b))

print(f"\nTotal unique products: {len(multiplication_table)}")
print(f"Products range: {min(multiplication_table.keys())} to {max(multiplication_table.keys())}")

# print first 5 and last 5 rows of the dataframe multiplication table
# df = pd.DataFrame(index=numbers, columns=numbers)
# for a in numbers:
#     for b in numbers:
#         df.at[a, b] = a * b
#
# df.head()  # Show first 5 rows
# df.tail()  # Show last 5 rows

# show last 5 rows 
print(dict(list(multiplication_table.items())[-5:]))

multiplication_table
```

## Condition 1: D doesn't know the numbers

**Logic**: The product must have multiple factorizations (multiple pairs yield the same product)

Player D knows the product but cannot uniquely determine the two numbers.

```python
def condition_1_d_not_knows(multiplication_table: dict) -> list:
    """
    Check if product has multiple factorizations (D cannot uniquely determine numbers).
    
    """
    table_cond_1 = []
    for product_i in multiplication_table:
        if len(multiplication_table[product_i]) > 1:
            table_cond_1.append(product_i)

    return table_cond_1

# Test case placeholder

len_2_plus_table = condition_1_d_not_knows(multiplication_table)
print(f"f: {len_2_plus_table}")

print(f"Product : {len_2_plus_table[3]}")
print(f"Factorizations: {multiplication_table[len_2_plus_table[3]]} \n")


#is 20 in the list of products with multiple factorizations?
print(f"Is 18 in the list condition 1? {18 in len_2_plus_table}")

# result = condition_1_d_not_knows(6, multiplication_table)
# print(f"Product 6 has multiple factorizations: {result}")
```

## Condition 2: C knows D doesn't know (C can deduce from the sum)

**Logic**: For the given sum, ALL possible pairs must have products with multiple factorizations

Player C, knowing the sum and that D doesn't know, can deduce this fact directly from the sum alone.

```python
# Generate sum table for numbers 2-99
sum_table = {}

for a in numbers:
    for b in numbers:
        if a <= b:  # Store only unique pairs (a <= b) to avoid duplicates
            sum_val = a + b
            if sum_val not in sum_table:
                sum_table[sum_val] = []
            sum_table[sum_val].append((a, b))

# show last 5 rows 
print(dict(list(sum_table.items())[-5:]))

def condition_2_c_knows_d_not_knows(sum_table, multiplication_table):
    table_cond_2 = []
    table_cond_1 = condition_1_d_not_knows(multiplication_table)

    for total_sum, pairs_in_sum in sum_table.items():
    # Check if ALL pairs have ambiguous products
        all_pairs_valid = True
        
        for a, b in pairs_in_sum:
            product = a * b
            if product not in table_cond_1 or len(multiplication_table[product]) < 2:
                all_pairs_valid = False
                break
        
        if all_pairs_valid:
            table_cond_2.append(total_sum)
  
    return table_cond_2


# Test it
valid_sums_cond2 = condition_2_c_knows_d_not_knows(sum_table, multiplication_table)
# print(f"Valid sums for Condition 2: {sorted(valid_sums_cond2)}")
print(f"Count: {len(valid_sums_cond2)}")

# is 18 
print(f"Is 18 in the list condition 2? {18 in valid_sums_cond2}")
print(f"Is 11 in the list condition 2? {11 in valid_sums_cond2}")

# Show examples
print("\n=== Examples ===")
for s in sorted(valid_sums_cond2)[:3]:
    print(f"Sum {s}:")
    for a, b in sum_table[s]:
        product = a * b
        print(f"  ({a}, {b}): {a}×{b}={product} → {len(multiplication_table[product])} factorizations")
```

## Solution Algorithm

The main algorithm follows these steps:

1. **Initialize**: Create multiplication table for all pairs (2-99)

2. **Find valid pairs** that satisfy ALL conditions:
   - Valid pair (a, b): `2 ≤ a ≤ b ≤ 99`
   - Product must satisfy Condition 1: has multiple factorizations
   - Product must satisfy Condition 3: unique after C's statement
   - Sum must satisfy Condition 2: all its pairs have ambiguous products
   - Sum must satisfy Condition 4: unique after D's statement

3. **Output**: All pairs (a, b) that satisfy the complete dialogue
